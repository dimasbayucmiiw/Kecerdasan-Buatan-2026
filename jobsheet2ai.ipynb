Hasil perkalian dalam matriks

```python
import numpy as np

A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

C = np.array([[2, 0],
              [1, 2]])

hasil_AB = np.dot(A, B)
hasil_BC = np.dot(B, C)

print("A x B =\n", hasil_AB)
print("B x C =\n", hasil_BC)
```

Menghitung nilai faktorial

```python
import math

faktorial_6 = math.factorial(6)

hasil = math.factorial(8) / math.factorial(5)

print("6! =", faktorial_6)
print("8! / 5! =", hasil)
```

Hasil operasi vektor

```python
import numpy as np

A = np.array([5, 3, 7])
B = np.array([12, -3, 1])

panjang_A = np.linalg.norm(A)
panjang_B = np.linalg.norm(B)

dot_product = np.dot(A, B)
cross_product = np.cross(A, B)

print("Panjang vektor A =", panjang_A)
print("Panjang vektor B =", panjang_B)
print("A . B =", dot_product)
print("A x B =", cross_product)
```
