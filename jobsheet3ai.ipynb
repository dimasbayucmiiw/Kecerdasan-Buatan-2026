Mendapatkan Nilai Mean

```python
import numpy as np

score = [90, 80, 70, 65, 110, 80, 83, 67, 94]
x = np.mean(score)
print("Hasil =", np.mean(score))
```

Mendapatkan nilai median

```python
import numpy as np

score = [90, 80, 70, 65, 110, 80, 83, 67, 94]
x = np.median(score)
print("Hasil =", np.median(score))
```

Mendapatkan nilai modus/mode

```python
from scipy import stats

score = [90, 80, 70, 65, 110, 80, 83, 67, 94]
x = stats.mode(score, keepdims=True)
print("Hasil =", stats.mode(score, keepdims=True))
```

Mendapatkan nilai range

```python
import numpy as np

data = [2, 3, 7, 8, 9, 12, 15, 20, 50, 42]

minimum = np.min(data)
maximum = np.max(data)
x = maximum - minimum
print("Hasil =", np.max(data) - np.min(data))
```

Mendapatkan nilai percentile

```python
import numpy as np

data = [90, 80, 70, 65, 110, 80, 83, 67, 94]
x = np.percentile(data, 90)
print("Hasil =", np.percentile(data, 90))
```

Mendapatkan Nilai Quartile dan Interquartile range

```python
import numpy as np

data = [3, 4, 4, 5, 5, 6, 6, 7, 8, 8, 9, 10]

Q1 = np.median(data[:int(len(data)/2)])
Q3 = np.median(data[int(len(data)/2):])
IQR = Q3 - Q1

print("Q1=", Q1)
print("Q3=", Q3)
print("IQR=", IQR)
```

Mendapatkan nilai variance dan standart deviation

```python
import numpy as np

data= [86, 87, 88, 86, 87, 85, 86]

mean = np.mean(data)
var = np.var(data)
std = np.std(data)

print("mean=", mean)
print("var=", var)
print("std=", std)
```

Analisis data ujian

```python
import numpy as np

data = [50,54,62,50,52,59,61,63,65,10,53,63,65,50,59,62,50,51,57,60,63,65,65,53,99]

sorted_data = sorted(data)

Q1 = np.percentile(sorted_data, 25)
Q3 = np.percentile(sorted_data, 75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outlier = [x for x in sorted_data if x < lower_bound or x > upper_bound]

print("Data urut =", sorted_data)
print("Q1 =", Q1)
print("Q3 =", Q3)
print("IQR =", IQR)
print("Lower Bound =", lower_bound)
print("Upper Bound =", upper_bound)
print("Outlier =", outlier)
```

Standart deviasi data berkelompok

```python
import numpy as np

interval = [(41,45), (46,50), (50,55), (56,60), (61,65), (66,70)]
frekuensi = [2,3,6,4,5,2]

titik_tengah = [(a+b)/2 for a,b in interval]

mean = np.sum(np.array(titik_tengah) * np.array(frekuensi)) / np.sum(frekuensi)

var = np.sum(frekuensi * (np.array(titik_tengah) - mean)**2) / np.sum(frekuensi)

std = np.sqrt(var)

print("Mean =", mean)
print("Standar Deviasi =", std)
```
